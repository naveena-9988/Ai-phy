```python
import numpy as np
import matplotlib.pyplot as plt

# values of n
N_values = [500, 1000, 2000, 5000, 10000, 15000, 20000, 50000, 100000]

for n in N_values:
    
    # 1) simulate two fair dice n times
    d1 = np.random.randint(1, 7, size=n)
    d2 = np.random.randint(1, 7, size=n)
    s = d1 + d2

    # 2) compute histogram using numpy
    h, h2 = np.histogram(s, bins=range(2, 14))   # bins 2..13
    
    # 3) plot histogram
    plt.figure(figsize=(6, 4))
    plt.bar(h2[:-1], h / n, width=0.8)
    
    plt.xticks(range(2, 13))
    plt.xlabel("Sum of Two Dice")
    plt.ylabel("Relative Frequency")
    plt.title(f"Histogram of Dice Sums (n = {n})")
    plt.ylim(0, 0.20)
    plt.grid(axis="y", linestyle=":", linewidth=0.5)
    
    plt.show()
```

```python
4. I observed 4.i observed Small n (e.g. 500)  bars are uneven; random sampling noise is large. Some sums will appear more/less often than theory predicts.

Moderate n (e.g. 5k20k)  the shape becomes triangular, peaks near 7, and valleys near 2 and 12.

Large n (50k100k)  the empirical frequencies closely match the theoretical probabilities listed in your table (1/36, 2/36, ..., 6/36 for sums 2..7 then symmetric).

Run the loop multiple times if you want to see how different random draws still converge as n grows.
```

```python
5.The mean  of the sum of two fair dice is 7.

With small samples, random variation causes observed averages and the shape of the distribution to deviate from the expected values  this is why small-n histograms look noisy and may show extreme deviations.

As sample size increases, sample averages and frequencies "regress" toward (i.e., move closer to) the true expected values  the empirical distribution converges to the theoretical distribution.

In plain terms: extreme or noisy results in small samples tend to cancel out when you collect more data, pulling the results back toward the expected  outcome.
```
